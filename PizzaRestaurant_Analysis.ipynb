## ðŸ› ï¸ Data Transformation


```python
pizza['Restaurant Name'] = pizza['Restaurant Name'].str.replace('â€™', "'", regex=False)
pizza['Restaurant Name'] = pizza['Restaurant Name'].str.strip()
```

```python
pizza['Restaurant Name'].value_counts()
```

```python
pizza[['City', 'State']] = pizza['Location'].str.split(', ', expand=True)
print(pizza)
```

```python
pizza['Delivery Time'] = pd.to_datetime(pizza['Delivery Time'])
pizza['Order Time'] = pd.to_datetime(pizza['Order Time'])
```

```python
pizza['Order Time'] = pd.to_datetime(pizza['Order Time']).dt.strftime('%I:%M')
pizza['Delivery Time'] = pd.to_datetime(pizza['Delivery Time']).dt.strftime('%I:%M')
```

```python
pizza
```

```python
pizza['Pizza Size'] = pizza['Pizza Size'].str.replace('XL', "X-Large", regex=False)
print(pizza['Pizza Size'])
```

## Modify The Part Of Time: Delivery, Delay And Estimated

```python
def smart_round(x):
    if x % 1 == 0.5:
        return x  
    else:
        return round(x)
pizza['Distance (km)'] = pizza['Distance (km)'].apply(smart_round)
```

```python
pizza['Distance (km)'].unique()
```

```python
pizza['Is Peak Hour'] = pizza['Is Peak Hour'].replace({'1': True, '0': False})
print(pizza['Is Peak Hour'])
```

```python
pizza['Is Weekend'] = pizza['Is Weekend'].replace({'1': True, '0': False})
print(pizza['Is Weekend'])
```

```python
avg_efficiency = pizza['Delivery Duration (min)'].sum() / pizza['Distance (km)'].sum()
pizza['Estimated Duration (min)'] = pizza['Distance (km)'] * avg_efficiency
```

```python
pizza['Estimated Duration (min)'].head(10)
```

```python
pizza['Estimated Duration (min)'] = pizza['Estimated Duration (min)'].round().astype(int)
pizza['Estimated Duration (min)']
```

```python
pizza['Delay (min)']=pizza['Delivery Duration (min)']-pizza['Estimated Duration (min)']
print(pizza)
```

```python
def classify_delay(val):
    if val < 0:
        return 'Early'
    elif val == 0:
        return 'On Time'
    else:
        return 'Delayed'

pizza['Delay (min)'] = pizza['Delay (min)'].astype(int)
pizza['Delay Status'] = pizza['Delay (min)'].apply(classify_delay)
```

## Change Some Names Of Our Values And Rearrange Columns

```python
def classify_density(val):
    if val < 0.6:
        return 'Low Density'
    elif val < 0.85:
        return 'Medium Density'
    else:
        return 'High Density'

pizza['Topping Density'] = pizza['Topping Density'].apply(classify_density)
```

```python
pizza
```

```python
def classify_complexity(val):
    if val <= 4:
        return 'Very Low'
    elif val <= 8:
        return 'Low'
    elif val <= 12:
        return 'Medium'
    else:
        return 'High'

pizza['Pizza Complexity'] = pizza['Pizza Complexity'].apply(classify_complexity)
```

## ðŸ“Š Exploratory Data Analysis (EDA)


```python
pizza
```

```python
# --- Monthly Sales (Quantity) ---
monthly_sales = pizza.groupby('Order Month')['Order ID'].count()
print(monthly_sales)
```

```python
# Number Of Orders In Each Restaurant 
pizza['Restaurant Name'].value_counts()
```

```python
# Number Of Orders By Each Type
pizza['Pizza Type'].value_counts()
```

```python
# Number Of Orders By Size
orders_per_size = pizza.groupby('Pizza Size')['Order ID'].nunique()
print(orders_per_size)
```

```python
# Relation Between Pizza Type And Size To Orders
relation_type_size= pizza.groupby(['Pizza Size','Pizza Type'])['Order ID'].nunique().unstack(fill_value=0)
print(relation_type_size)
```

```python
# Top 10 states And Cities That Order Pizza
top10_City=pizza.groupby('City')['Order ID'].nunique().sort_values(ascending=False).head(10)
top10_State=pizza.groupby('State')['Order ID'].nunique().sort_values(ascending=False).head(10)
print(top10_City)
print(top10_State)
```

```python
# Number Of Orders In Each Month
Orders_per_month=pizza.groupby('Order Month')['Order ID'].nunique().sort_values(ascending=False)
print(Orders_per_month)
```

```python
# Impact of Pizza Complexity and Topping on Orders
orders_by_complexity_density = pizza.groupby(['Pizza Complexity', 'Topping Density'])['Order ID'].nunique().reset_index(name='Order Count').sort_values('Order Count', ascending=False)
print(orders_by_complexity_density)
```

```python
# All Details About Pizza Attributes With Pizza Type

complexity_dist = pizza['Pizza Complexity'].value_counts(normalize=True) * 100
print("Pizza Complexity Distribution (%):\n", complexity_dist)

density_dist = pizza['Topping Density'].value_counts(normalize=True) * 100
print("\nTopping Density Distribution (%):\n", density_dist)

toppings_dist = pizza['Toppings Count'].value_counts(normalize=True) * 100
print("\nToppings Count Distribution (%):\n", toppings_dist)

complexity_by_type = pizza.groupby(['Pizza Type', 'Pizza Complexity'])['Order ID'].nunique().unstack(fill_value=0)
print("\nOrders by Pizza Type and Complexity:\n", complexity_by_type)

density_by_type = pizza.groupby(['Pizza Type', 'Topping Density'])['Order ID'].nunique().unstack(fill_value=0)
print("\nOrders by Pizza Type and Density:\n", density_by_type)

toppings_by_type = pizza.groupby(['Pizza Type', 'Toppings Count'])['Order ID'].nunique().unstack(fill_value=0)
print("\nOrders by Pizza Type and Toppings Count:\n", toppings_by_type)

most_common_complexity = complexity_by_type.idxmax(axis=1)
print("\nMost Common Complexity per Pizza Type:\n", most_common_complexity)

most_common_density = density_by_type.idxmax(axis=1)
print("\nMost Common Density per Pizza Type:\n", most_common_density)

most_common_toppings = toppings_by_type.idxmax(axis=1)
print("\nMost Common toppings per Pizza Type:\n", most_common_toppings)
```

```python
# Number Of Orders By Time
Orders_by_time=pizza.groupby('Order Time')['Order ID'].nunique().sort_values(ascending=False)
print(Orders_by_time)
```

```python
# Relation Between Pizza Complexity and Delivery Time
avg_delivery_by_complexity = pizza.groupby('Pizza Complexity')['Delivery Duration (min)'].mean().sort_values(ascending=False)
print(avg_delivery_by_complexity)
```

```python
# Impact Of Weekend , Peak Hour on Delivery Time And Orders

# ØªØ£Ø«ÙŠØ± Ù†Ù‡Ø§ÙŠØ© Ø§Ù„Ø£Ø³Ø¨ÙˆØ¹ Ø¹Ù„Ù‰ Ù…Ø¯Ø© Ø§Ù„ØªÙˆØµÙŠÙ„ ÙˆØ¹Ø¯Ø¯ Ø§Ù„Ø·Ù„Ø¨Ø§Øª
weekend_impact = pizza.groupby('Is Weekend').agg(
    avg_delivery=('Delivery Duration (min)', 'mean'),
    order_count=('Order ID', 'nunique')
)
print("Impact of Weekend:\n", weekend_impact)

# ØªØ£Ø«ÙŠØ± Ø£ÙˆÙ‚Ø§Øª Ø§Ù„Ø°Ø±ÙˆØ© Ø¹Ù„Ù‰ Ù…Ø¯Ø© Ø§Ù„ØªÙˆØµÙŠÙ„ ÙˆØ¹Ø¯Ø¯ Ø§Ù„Ø·Ù„Ø¨Ø§Øª
peak_impact = pizza.groupby('Is Peak Hour').agg(
    avg_delivery=('Delivery Duration (min)', 'mean'),
    order_count=('Order ID', 'nunique')
)
print("\nImpact of Peak Hour:\n", peak_impact)
```
